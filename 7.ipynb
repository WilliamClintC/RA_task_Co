# PDF Text Extraction

1. Read a PDF file
2. Extract all text content
3. Find specific lines containing "Engineer's Estimate:" and "Bidder" followed by numbers

```python
# Import required libraries
import PyPDF2
import re
import os
import pandas as pd
```

```python
# Path to the PDF file
pdf_path = r"C:\Users\clint\Desktop\RA Task\Downloads\180003bidtab.pdf"

# Verify file exists
if not os.path.exists(pdf_path):
    print(f"File not found: {pdf_path}")
else:
    print(f"File found: {pdf_path}")
    
    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        # Create PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Get number of pages
        num_pages = len(pdf_reader.pages)
        print(f"Number of pages: {num_pages}")
        
        # Extract text from all pages
        all_text = ""
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            all_text += page.extract_text()
        
        # Display first 1000 characters to see sample content
        print("\nSample content (first 1000 characters):\n")
        print(all_text[:1000])
```

```python
# Extract specific lines using regular expressions
engineers_estimate_pattern = r"Engineer's Estimate:.*"
bidder_pattern = r"Bidder \d+.*"

# Find all matches
engineers_estimate = re.findall(engineers_estimate_pattern, all_text)
bidders = re.findall(bidder_pattern, all_text)

# Display results
print("Engineer's Estimate:")
for estimate in engineers_estimate:
    print(f"  {estimate}")

print("\nBidders:")
for bidder in bidders:
    print(f"  {bidder}")
```

```python
csv_path = r"C:\Users\clint\Desktop\RA Task\5.csv"
df_csv = pd.read_csv(csv_path)
df_csv
```

```python
excel_path = r"C:\Users\clint\Desktop\RA Task\Ohio_2018_Resurfacing PRR.xlsx"
df_excel = pd.read_excel(excel_path)
df_excel.head()
```

```python
# Create a mapping from PID to Project Num in df_excel
pid_to_project_num = dict(zip(df_excel['PID'], df_excel['Project Num']))

# Create a new column in df_csv using the mapping
df_csv['Project Num'] = df_csv['project_id'].map(pid_to_project_num)

# Display the first few rows to verify
df_csv.head()
```

```python
# Extract the Project Num directly from the PDF filename
project_num_match = re.search(r'([0-9]+)bidtab\.pdf', pdf_path)
if project_num_match:
    project_num_str = project_num_match.group(1)  # Keep original as string
    project_num = int(project_num_str)  # Convert to integer for matching
    print(f"Extracted Project Num: {project_num} (type: {type(project_num)})")
    
    # Extract the dollar amount from the Engineer's Estimate
    if engineers_estimate:
        dollar_amount = re.search(r'\$([0-9,]+\.[0-9]+)', engineers_estimate[0])
        if dollar_amount:
            # Convert the amount to a float (removing commas) and then to millions
            dollar_amount_str = dollar_amount.group(1).replace(',', '')
            amount_in_millions = float(dollar_amount_str) / 1_000_000
            print(f"Engineer's Estimate in millions: {amount_in_millions}")
            
            # Update the DataFrame where Project Num matches
            mask = df_csv['Project Num'] == project_num  # Now comparing integers to integers
            if mask.any():
                df_csv.loc[mask, 'eng_estimate_mils'] = amount_in_millions
                print(f"Updated eng_estimate_mils for Project Num {project_num}")
                # Display the updated row
                print(df_csv.loc[mask])
            else:
                print(f"No matching Project Num {project_num} found in the DataFrame")
                # Double-check with string comparison as a fallback
                alt_mask = df_csv['Project Num'].astype(str) == project_num_str
                if alt_mask.any():
                    print(f"Found a match when comparing as strings!")
                    df_csv.loc[alt_mask, 'eng_estimate_mils'] = amount_in_millions
                    print(f"Updated eng_estimate_mils using string comparison for Project Num {project_num_str}")
                    print(df_csv.loc[alt_mask])
        else:
            print("Could not extract dollar amount from Engineer's Estimate")
    else:
        print("Engineer's Estimate not found in the PDF")
    
    # Extract the number of bidders by finding the highest bidder number
    if bidders:
        # Extract the bidder numbers from the bidder strings
        bidder_numbers = []
        for bidder in bidders:
            bidder_num_match = re.search(r'Bidder (\d+)', bidder)
            if bidder_num_match:
                bidder_numbers.append(int(bidder_num_match.group(1)))
        
        if bidder_numbers:
            num_bidders = max(bidder_numbers)  # The highest bidder number is the count of bidders
            print(f"Number of bidders: {num_bidders}")
            
            # Update the DataFrame with the number of bidders
            mask = df_csv['Project Num'] == project_num
            if mask.any():
                df_csv.loc[mask, 'num_bidders'] = num_bidders
                print(f"Updated num_bidders for Project Num {project_num}")
                # Display the updated row
                print(df_csv.loc[mask])
            else:
                # Try string comparison as a fallback
                alt_mask = df_csv['Project Num'].astype(str) == project_num_str
                if alt_mask.any():
                    df_csv.loc[alt_mask, 'num_bidders'] = num_bidders
                    print(f"Updated num_bidders using string comparison for Project Num {project_num_str}")
                    print(df_csv.loc[alt_mask])
                else:
                    print(f"No matching Project Num {project_num} found in the DataFrame for updating num_bidders")
        else:
            print("Could not extract bidder numbers from the bidder strings")
    else:
        print("No bidders found in the PDF")
else:
    print("Could not extract Project Num from the filename")

# Show the first few rows of the updated DataFrame
df_csv
```

```python
# Function to process a single PDF and update the DataFrame
def process_bidtab_pdf(pdf_path, df):
    print(f"Processing {os.path.basename(pdf_path)}...")
    
    # Open the PDF and extract text
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        all_text = ""
        for page in pdf_reader.pages:
            all_text += page.extract_text()
    
    # Extract Engineer's Estimate and bidders
    engineers_estimate = re.findall(r"Engineer's Estimate:.*", all_text)
    bidders = re.findall(r"Bidder \d+.*", all_text)
    
    # Extract Project Num from filename
    project_num_match = re.search(r'([0-9]+)bidtab\.pdf', pdf_path)
    if not project_num_match:
        print(f"  Could not extract Project Num from filename: {os.path.basename(pdf_path)}")
        return
    
    project_num_str = project_num_match.group(1)
    project_num = int(project_num_str)
    print(f"  Project Num: {project_num}")
    
    # Process Engineer's Estimate
    eng_estimate_mils = None
    if engineers_estimate:
        dollar_amount = re.search(r'\$([0-9,]+\.[0-9]+)', engineers_estimate[0])
        if dollar_amount:
            dollar_amount_str = dollar_amount.group(1).replace(',', '')
            eng_estimate_mils = float(dollar_amount_str) / 1_000_000
            print(f"  Engineer's Estimate: ${eng_estimate_mils:.2f} million")
        else:
            print("  Could not extract dollar amount from Engineer's Estimate")
    else:
        print("  Engineer's Estimate not found in the PDF")
    
    # Process number of bidders
    num_bidders = None
    if bidders:
        bidder_numbers = []
        for bidder in bidders:
            bidder_num_match = re.search(r'Bidder (\d+)', bidder)
            if bidder_num_match:
                bidder_numbers.append(int(bidder_num_match.group(1)))
        
        if bidder_numbers:
            num_bidders = max(bidder_numbers)
            print(f"  Number of bidders: {num_bidders}")
        else:
            print("  Could not extract bidder numbers from bidder strings")
    else:
        print("  No bidders found in the PDF")
    
    # Update the DataFrame
    mask = df['Project Num'] == project_num
    alt_mask = df['Project Num'].astype(str) == project_num_str
    
    if mask.any():
        # Update using integer comparison
        if eng_estimate_mils is not None:
            df.loc[mask, 'eng_estimate_mils'] = eng_estimate_mils
        if num_bidders is not None:
            df.loc[mask, 'num_bidders'] = num_bidders
        print(f"  Updated DataFrame for Project Num {project_num}")
        return True
    elif alt_mask.any():
        # Update using string comparison
        if eng_estimate_mils is not None:
            df.loc[alt_mask, 'eng_estimate_mils'] = eng_estimate_mils
        if num_bidders is not None:
            df.loc[alt_mask, 'num_bidders'] = num_bidders
        print(f"  Updated DataFrame for Project Num {project_num_str} (string comparison)")
        return True
    else:
        print(f"  No matching Project Num {project_num} found in the DataFrame")
        return False
```

```python
# Get all bidtab PDF files in the Downloads directory
downloads_dir = r"C:\Users\clint\Desktop\RA Task\Downloads"
bidtab_pdfs = [os.path.join(downloads_dir, f) for f in os.listdir(downloads_dir) 
              if f.endswith('bidtab.pdf')]

print(f"Found {len(bidtab_pdfs)} bidtab PDF files")

# Process each PDF file
successful_updates = 0
for pdf_path in bidtab_pdfs:
    if process_bidtab_pdf(pdf_path, df_csv):
        successful_updates += 1
    print("---")

print(f"Successfully updated data for {successful_updates} out of {len(bidtab_pdfs)} PDFs")

# Display the columns with the updates
df_csv[['Project Num', 'eng_estimate_mils', 'num_bidders']]
```
