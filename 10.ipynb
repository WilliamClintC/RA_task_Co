```python
df2 = pd.read_csv(r'C:\Users\clint\Desktop\RA Task\7.csv')
df2
```

```python
import re

def clean_bidder_name(name):
    # Convert to title case
    name = name.title()
    # Remove common company suffixes for professionalism
    name = re.sub(r'\b(Inc|Llc|Corp|Corporation|Limited|Ltd|Co|Company|Incorporated)\b[., ]*', '', name, flags=re.IGNORECASE)
    # Remove extra spaces and commas
    name = re.sub(r'\s{2,}', ' ', name).strip()
    name = re.sub(r',\s*,', ',', name)
    name = name.strip(', ')
    return name

# Ensure 'bidders_list' column exists in df2
if 'bidders_list' not in df2.columns:
    df2['bidders_list'] = ''

for idx, row in df.iterrows():
    project_number = row['project_number']
    bidder_name = clean_bidder_name(str(row['bidder_name']))
    # Find matching rows in df2
    match = df2['Project Num'] == project_number
    for i in df2[match].index:
        current = df2.at[i, 'bidders_list']
        if pd.isna(current) or current == '':
            df2.at[i, 'bidders_list'] = bidder_name
        else:
            # Split, clean, and rejoin to avoid duplicates and ensure formatting
            names = [clean_bidder_name(n) for n in current.split(',') if n.strip()]
            if bidder_name not in names:
                names.append(bidder_name)
            df2.at[i, 'bidders_list'] = ', '.join(names)

df2
```

```python
# Check robustness: Compare number of bidders in 'bidders_list' to 'num_bidders' in df2

def count_bidders(bidders):
    if pd.isna(bidders) or bidders.strip() == '':
        return 0
    return len([b.strip() for b in bidders.split(',') if b.strip()])

# Calculate the number of bidders in 'bidders_list'
df2['bidders_count'] = df2['bidders_list'].apply(count_bidders)

# Find rows where the count does not match 'num_bidders'
mismatches = df2[df2['bidders_count'] != df2['num_bidders']]

if mismatches.empty:
    print('All entries are consistent: number of bidders matches num_bidders.')
else:
    print('Rows with mismatched bidder counts:')
    display(mismatches[['Project Num', 'bidders_list', 'bidders_count', 'num_bidders']])
```
