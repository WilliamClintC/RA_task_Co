```python
import pandas as pd

df = pd.read_csv(r'C:\Users\clint\Desktop\RA Task\13.csv')
df
```

```python
df = df.drop(index=0).reset_index(drop=True)
df
```

```python
df['lanes'] = df['lanes'].astype('Int64')
df['num_bidders'] = df['num_bidders'].astype('Int64')
df.dtypes[['lanes', 'num_bidders']]
```

```python
df = df.drop(columns=['all_routes','bidders_count','cost_overrun_pct','Project Num'])
df.head()
```

```python
# Read the Excel file to see all available columns
ohio_excel = pd.read_excel(r'C:\Users\clint\Desktop\RA Task\Ohio_projects_collected.xlsx')
print("Columns in Ohio_projects_collected.xlsx:")
print(ohio_excel.columns.tolist())
print(f"\nShape: {ohio_excel.shape}")
ohio_excel.head()
```

```python
# Create final_output_specific with only the specified columns
specific_columns = [
    'project_id', 'route', 'mileage', 'lanes', 'project_duration_days',
    'eng_estimate_mils', 'win_bid_mils', 'cost_mils', 'num_bidders', 'bidders_list'
]

# Check which columns exist in our dataframe
existing_specific_cols = [col for col in specific_columns if col in df.columns]
print(f"Available specific columns: {existing_specific_cols}")

# Create the specific output
final_output_specific = df[existing_specific_cols].copy()

# Create final_output_complete with all columns from Ohio_projects_collected.xlsx
complete_columns = ohio_excel.columns.tolist()
existing_complete_cols = [col for col in complete_columns if col in df.columns]
print(f"Available complete columns: {existing_complete_cols}")

# Create the complete output
final_output_complete = df[existing_complete_cols].copy()

# Save both outputs
final_output_specific.to_csv('final_output_specific.csv', index=False)
final_output_complete.to_csv('final_output_complete.csv', index=False)

print(f"\nfinal_output_specific shape: {final_output_specific.shape}")
print(f"final_output_complete shape: {final_output_complete.shape}")
print("\nBoth files saved successfully!")

# Display first few rows of each
print("\n=== FINAL OUTPUT SPECIFIC ===")
print(final_output_specific.head())
print("\n=== FINAL OUTPUT COMPLETE ===")
print(final_output_complete.head())
```

```python

# Check for NaN values in both datasets
print("\n" + "="*50)
print("CHECKING FOR NaN VALUES")
print("="*50)

print("\n=== NaN COUNT IN FINAL_OUTPUT_SPECIFIC ===")
nan_count_specific = final_output_specific.isnull().sum()
print(nan_count_specific)
print(f"\nTotal NaN values in specific output: {nan_count_specific.sum()}")

print("\n=== NaN COUNT IN FINAL_OUTPUT_COMPLETE ===")
nan_count_complete = final_output_complete.isnull().sum()
print(nan_count_complete)
print(f"\nTotal NaN values in complete output: {nan_count_complete.sum()}")

# Show rows with NaN values in specific columns
print("\n=== ROWS WITH NaN IN SPECIFIC OUTPUT ===")
if final_output_specific.isnull().any().any():
    rows_with_nan = final_output_specific[final_output_specific.isnull().any(axis=1)]
    print(f"Number of rows with NaN: {len(rows_with_nan)}")
    print(rows_with_nan[['project_id', 'route', 'mileage']].head(10))
else:
    print("No NaN values found in specific output!")

# Check which specific columns have NaN
print("\n=== COLUMNS WITH NaN IN SPECIFIC OUTPUT ===")
columns_with_nan = nan_count_specific[nan_count_specific > 0]
if len(columns_with_nan) > 0:
    for col, count in columns_with_nan.items():
        print(f"{col}: {count} NaN values ({count/len(final_output_specific)*100:.1f}%)")
else:
    print("No columns have NaN values in specific output!")
```
